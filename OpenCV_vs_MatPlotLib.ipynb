```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.image import imread

flower_plt = plt.imread('flower.jpg')
print(flower_plt.dtype)
print(flower_plt.shape)
flower_plt = np.array(flower_plt)

flower_cv2 = cv2.imread('flower.jpg')
print(flower_cv2.dtype)
print(flower_cv2.shape)
flower_cv2 = np.array(flower_cv2)

# Create a 2x2 grid for comparison
plt.figure(figsize=(10, 5))

# Original colored image
plt.subplot(1, 2, 1)
plt.imshow(flower_plt)
plt.title("Flower_plt")

# Grayscale image
plt.subplot(1, 2, 2)
plt.imshow(flower_cv2)
plt.title("Flower_cv2")

# Show the grid view
plt.tight_layout()
plt.show()
```

In this comparison, it is important to note that OpenCV reads images with a BGR channel ordering, while Matplotlib reads images with an RGB format.
